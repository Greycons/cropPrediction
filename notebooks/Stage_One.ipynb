```python
numeric_cols = data.select_dtypes(include=np.number).columns
```

```python
plt.figure(figsize=(15, 10))
data[numeric_cols].hist(bins=20, figsize=(15, 10), color='skyblue', edgecolor='black')
plt.suptitle("Numeric Feature Distributions", fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

params = ['Groundwater_pH', 'EC_Groundwater (µS/cm)', 'Hardness_Groundwater (mg/l)', 'Nitrate_Groundwater (mg/l)', 
           'Soil_pH', 'Soil_Organic_Carbon', 'Soil_Nitrogen', 'Soil_Phosphorus', 'Soil_Potassium']

plt.figure(figsize=(15, 10))
for i, col in enumerate(params, 1):
    plt.subplot(3, 3, i)
    sns.histplot(data[col], kde=True, color='teal')
    plt.title(col.replace('_', ' '))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=data, x='Rainfall_mm', y='Crop_Yield', hue='District')
plt.title('Rainfall vs Crop Yield by District')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Crop Yield')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=data, x='Soil_pH', y='Soil_Organic_Carbon', hue='Crop', style='State')
plt.title('Soil pH vs Organic Carbon')
plt.show()
```

```python
avg_yield = data.groupby('District')['Crop_Yield'].mean().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=avg_yield.index, y=avg_yield.values, palette='viridis')
plt.title('Average Crop Yield by District')
plt.xticks(rotation=45)
plt.ylabel('Average Yield')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Crop', y='Crop_Yield', data=data, palette='coolwarm')
plt.title('Crop-wise Yield Distribution')
plt.xlabel('Crop')
plt.ylabel('Crop Yield')
plt.show()
```

```python
sns.pairplot(data[['Soil_pH', 'Soil_Nitrogen', 'Rainfall_mm', 'Crop_Yield']], diag_kind='kde', hue='Crop')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['rainfall_mm'], data['soil_nitrogen'], data['crop_yield'], c='green')
ax.set_xlabel('rainfall (mm)')
ax.set_ylabel('soil nitrogen')
ax.set_zlabel('crop yield')
plt.title('3D Interaction: Rainfall, Soil Nitrogen, and Yield')
plt.show()
```
